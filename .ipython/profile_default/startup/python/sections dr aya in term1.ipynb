```python
hotels = {}

#step1
try:
    with open("hotels.txt","r") as file:
        for line in file:
            names,rooms = line.strip().split(",")
            hotels[names.strip()]=int(rooms.strip())
except FileNotFoundError:
    pass

#step2
hotels["e.g."] = 6
hotels["Hilton"] = 4
hotels["four seasons"] = 3

#step3
print("available rooms :")
for names,rooms in hotels.items():
    print(f"{names}:{rooms}")
    
#step4
action = str("choose action (book,cancel)")
name_hotel = str("enter the hotel name")
if name_hotel not in hotels:
    print("hotel not found")
else:
    if action == "book":
        if hotels[name_hotel]>0:
            hotels[name_hotel]-= 1
            print("you booked")
            action = "booked"
        else:
            print("hotel is full")
    elif action == "cancel":
        hotels[name_hotel]+= 1
        print("you cancelled")
        action = "cancelled"
    else:
        print("action is not found")
        
#step5
with open("ticket.txt","a") as file:
    file.write(f"{action}:{name_hotel}")
    
#step6
with open("hotel.txt","w") as file:
    for names,rooms in hotels.items():
        file.write(f"{names}:{rooms}")
```

```python
movies = {}

#step1
try:
    with open("movies.txt","r")as file:
        for line in file:
            if "," not in line:
                
                continue   # يتجاهل السطر الغلط
                movie,seats = line.strip().split(",")
                movies[movie.strip()] = int(seats.strip())
except FileNotFoundError:
    pass

#step2
movies["midterm"] = 65
movies["wa lana fe el khaial hob"] = 74
movies["om kolsoom"] = 34

#step3
for movie,seats in movies.items():
    print(f"{movie}:{seats}")
        
#step4
action = str(input("select action(book or cancell)"))
movie = str(input("enter name of film"))
number_of_seats = int(input("enter number of seats"))
if movie in movies:
   if action == "book":
       if movies[movie] >= number_of_seats:
          movies[movie]-= number_of_seats
          print("you book the seats")
          action = "booked"
       else:
           print("not available")
   elif action == "cancel":
       movies[movie]+= number_of_seats
       print("you cancelled")
       action = "cancelled"
   else:
        print("enter true action")
else:
    print("movie is not available")
    
#step5
with open("tickets.txt","a",encoding="utf-8")as file:
     file.write(f"{action}:{movie}\n")
     
#step6
with open("movies.txt","w",encoding="utf-8")as file:
    for movie,seats in movies.items():
        file.write(f"{movie}:{seats}")       
    
```

```python
books = {}

#step1
try:
    with open("books.txt","r")as file:
        for line in file:
            names,copies = line.strip().split(":")
            books[names.strip()]=int(copies.strip())
except FileNotFoundError:
    pass

#step2
books["رسايل من القران"] = 4
books["جلسات نفسيه"] = 6
books["روايه ارض زيكولا"] = 2

#step3
for names,copies in books.items():
    print(f"{names}:{copies}")
    
#step4
action = str(input("borrow/return"))
book_name = str(input("enter the book name"))
if action == "borrow":
    if  book_name in books:
        if books[book_name] >= 1:
            print("you borrow the book")
            books[book_name]-= 1
            action = "booked"
        else:
            print("this book not in books now")
    else:
        print("this book not found")
elif action == "return":
    books[book_name]+= 1
    print("you return the book")
    action = "returned"
else:
    print("enter true action")
    
#step5
with open("log.txt","a")as file:
    file.write(f"{action}:{book_name}\n")
    
#step6
with open("books.txt","w")as file:
    for names,copies in books.items():
        file.write(f"{names}:{copies}\n")
        

            
```
