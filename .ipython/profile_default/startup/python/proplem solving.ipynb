```python
def remove_word(string):
    content = string.split()
    words = []
    for word in content:
        if word not in words:
        #if len(word) != 1 and word not in words:
            words.append(word)
    return " ".join(words)

print(remove_word("i love python i love programming programming"))

###############
def remove_word(sentense):
    return " ".join(list(dict.fromkeys(sentense)))
print(remove_word("i loev python i love python"))

    
```

```python
def add_commas_and_underscores(number):
    s = str(number)
    #for num in len(number):
    if len(s) == 7 or len(s) == 8 or len(s) == 9:
        return s[:-6] + "," + s[-6:-3] + "_" + s[-3:]
    elif len(s) == 4 or 5 or 6:
        return s[:-3] + "_" + s[-3:]
    elif len(s) == 3:
        return s
    
print(add_commas_and_underscores(12345679))
        
            
```

```python
def missing_letter(letters):
    letters1 = "abcdefghijklmnopqrstvwxyz"
    start = letters1.index(letters[0])
    end = letters1.index(letters[-1])
    letter =[]
    for i in letters1[start:end]:
        if i not in letters:
            letter.append(i)
    return letter
       
print(missing_letter("abdefg"))
```

```python
def words(word):
    w = []
    for word1 in word.split():
        if word1 not in w:
            w.append(word1)
    return "".join(w)
print(words("i i love love python python"))
```

```python
def reverse_list(number):
    content = str(number)
    list = []
    for char in content:
        list.append(int(char))
        list.reverse()
    return list
print(reverse_list(123456789))    
```
