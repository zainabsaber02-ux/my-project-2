#use python lists and functions to manage student grades .

add_student() - add student and grade 

display_students() - show all students

calculate_average() - show average grade

hieghest_-grade() - show top student

menu - choose action  


```python
students = []
def add_student():
    name = str(input("enter the name of student :"))
    grade = int(input("enter the grade of student :"))
    s = [name,grade]
    students.append(s)
       
def display_students():
    if len(students) == 0:
        print("no students")
    else:
        for s in students:
            print(f"name:{s[0]},grade:{s[1]}")
            
def calculate_average():
    if len(students) == 0:
        print("no average")
        return
    total = 0
    for s in students:
        total += s[1]
    average = total/len(students)
    print(f"average of students = {average}")
        
def highest_grade():
    if len(students) == 0:
        print("no student in students")
    else:
        max_degree = students[0][1]
        top_student = students[0][0]
        for s in students:
           if s[1] > max_degree:
            max_degree = s[1]
            top_student = s[0]
        print(f"top student:{top_student}:max_degree:{max_degree}")
           

while True:
    print("1:add_student")
    print("2:display_student")
    print("3:calculate_average")
    print("4:higest_grade")
    print("5:end")
    
    choice = str(input("enter your choice"))
    if choice == "1":
        add_student()
    elif choice == "2":
        display_students()
    elif choice == "3":
        calculate_average()
    elif choice == "4":
        highest_grade()
    elif choice == "5":
        print("programm end")
        break
    else:
        print("invaild")
        
```

```python
accounts = {}
def create_acount():
    acc_number = int(input("enter the account number"))
    
    if acc_number in accounts:
        print("acc_number already  exsist")
        return
    else:
        name = str(input("enter the name"))
        balance = int(input("enter the balance"))
        accounts[acc_number] = {'name': name, 'balance': balance}
        print("account created !")
        
def deposit():
    acc_number = int(input("enter the account number"))
    if acc_number not in accounts:
        print("this account not available")
    else:
    
        deposit1 = int(input("enter deposit"))
        accounts[acc_number]["balance"] += deposit1
        
def withdraw():
    acc_number = int(input("enter the acc_number"))
    if acc_number not in accounts:
        print("this account not in accounts")
    else:
        with_draw1 = int(input("enter withdraw amount"))
        if with_draw1 > accounts[acc_number]["balance"]:
            print("amount not available")
        else:
            accounts[acc_number]["balance"] -= with_draw1
            
def check_balance():
    acc_number = int(input("enter the acc-number"))
    if acc_number not in accounts:
        print("not available")
    else:
        
        print(accounts[acc_number]["balance"])
        
def display_accounts():
    for acc_number,data in accounts.items():
        print(acc_number,data)
while True:
    print("1:create-account")
    print("2:despoit")
    print("3:with_draw")
    print("4:check_balance")
    print("5:end")
    choice = input("enter your choice")
    if choice == "1":
        create_acount()
    elif choice == "2":
        deposit()
    elif choice == "3":
    
        withdraw()
    elif choice == "4":
        check_balance()
    elif choice == "5":
        print("programm end")
        break
    else:
        print("invaild")
          
            
```

student management system create a python program to manage student data with a menu:

add_student (name+marks for math,english,science)

show_students(name,marks,average,pass/fail)

search_student()

delete_student()

-exit

#use a list to store students and a dictionary for each student

#use functions,loops,and conditions

```python
students = []
def add_student():
    name = str(input("enter the name"))
    marks = {}
    
    subjects = ["math","english","science"]
    for sub in subjects:
        mark = float(input("enter the mark"))
        marks[sub] = mark
        
    student = {
       "name":name,
       "marks":marks
    }
    students.append(student)
        
def show_students():
    
    if not students:
        print("name not found")
        return
    for s in students:
        marks = s["marks"]
        aveg = sum(marks.values())/len(marks)
        status = "pass" if aveg > 50 else "fail"
        print(f"name:{s["name"]},marks:{marks},average:{aveg},status:{status}")
        
def search_student():
    name = str(input("ente the name"))
    for s in students:
       if name == s["name"]:
          print(f"name:{name},marks:{s["marks"]}")
    else:
        print("name not found")
            
def delete_student():
    name = str(input("enter the name"))
    if name not in students:
        print("student not found")
    else:
        for s in students:
           students.remove(s[name])
        
while True:
    print("1-add_student")
    print("2-show_students")
    print("3-search_student")
    print("4-delete-student")
    print("5-exit")
    choice = input("enter your choice")
    if choice == "1":
        add_student()
    elif choice == "2":
        show_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
       delete_student()
    elif choice == "5":
        print("programm end")
        break
    else:
        print("invaild")
```

your program should allow the user to perform the following operations:

add a new student (enter name and 3 marks) display all students calculate and display the average marks for each student save student data to a file named students.txt load and display student data from the file 

the program must use a menu_driven system that keeps running until the user chooses to exit

```python
students = []
def add_student():
    name = str(input("enter the name"))
    marks = []
    
    for mark in range(3):
        mark = float(input("enter the mark"))
        marks.append(mark)
        
    student = {
       "name":name,
       "marks":marks
    }
    students.append(student)
        
def show_students():
    
    if len(students) == 0:
        print("student not found")
        return
    for s in students:
        marks = s["marks"]
        print(f"name:{s['name']},marks:{marks}")
        
def calculate_average():
    for s in students:
        aveg = sum(s["marks"])/len(s["marks"])
        print(f"{s['name']},{aveg}")
    
def save_to_file():
    with open("students.txt","w")as file:
        for s in students:
            file.write(f"{s['name']},{s['marks']}\n")
    print("data is saved")
    
def load_data():
    with open("students.txt","r")as file:
        content = file.readlines()
        for line in content:
            print(line.strip())
            
while True:
    print("1-add_student")
    print("2-show_students")
    print("3-calculate_average")
    print("4-save_to_file")
    print("5-load_data")
    print("6-exit")
    choice = input("enter your choice")
    if choice == "1":
        add_student()
    elif choice == "2":
        show_students()
    elif choice == "3":
        calculate_average()
    elif choice == "4":
        save_to_file()
    elif choice == "5":
        load_data()
    elif choice == "6":
        print("programm end")
        break
    else:
        print("invaild")
           
        
```
